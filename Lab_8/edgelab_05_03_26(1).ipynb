# Detecting edges with Canny

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

input_video_path = '/content/262189_small.mp4'
output_video_path = '/content/canny_edges_video.avi' # Changed to .avi extension

cap = cv2.VideoCapture(input_video_path)

if not cap.isOpened():
    print(f"Error: Could not open video file at {input_video_path}")
else:
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    fourcc = cv2.VideoWriter_fourcc(*'XVID')

    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height), isColor=False)

    if not out.isOpened():
        print("Error: Could not create video writer.")
    else:
        print(f"Processing video and saving to {output_video_path}...")
        frame_count = 0
        first_frame_displayed = False

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

            canny_edges = cv2.Canny(gray_frame, 100, 200)

            if not first_frame_displayed:
                print("\nPreview of the first processed frame:")
                cv2_imshow(canny_edges)
                first_frame_displayed = True

            out.write(canny_edges)

            frame_count += 1
            if frame_count % 100 == 0:
                print(f"Processed {frame_count} frames...")

        print(f"\nFinished processing {frame_count} frames. Video saved as {output_video_path}")

    cap.release()
    out.release()
    cv2.destroyAllWindows()
```
